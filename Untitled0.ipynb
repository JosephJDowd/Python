```python
bearded_dragon = {
    "species": "reptile",
    "legs": 4,
    "favorite_food": "crickets",
    "likes_heat": True,
    "owners": [
        {
            "name": "Sebastien",
            "occupation": "Online instructor",
            "level_of_love_for_animal": "10"
        },
        {
            "name": "Celine Dion",
            "occupation": "Being rich",
            "level_of_love_for_animal": "1"
        }
    ]
}

unsatified_cat = {
    "species": "mammal",
    "legs": 4,
    "favorite_food": "uncertain",
}

animals = [bearded_dragon, unsatified_cat]
for animal in animals:
  print(animal)
  for key, value in animal.items():
    print(f"{key.capitalize()}: {value}")
```

```python
my_list = [10, 20, 30, 40, 50]
print("Initial list:", my_list)
```

```python
print("Third item: ", my_list[2])

list_length = len(my_list)
print("Length of the list:", list_length)

if list_length == 0:
  print("list is empty")
else:
  print("list is not empty")
```

```python
condition = "snake"

match condition:
  case "dog":
    print("This is not a cat!")
  case "cat":
    print("This IS a cat")
  case _:
    print("This is default behaviour")

if condition == "dog" and type(condition) == str: # or and group them with parenthesis
  print("The string contained the word dog")
```
